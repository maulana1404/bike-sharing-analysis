## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** Bagaimana perbandingan rata-rata jumlah penyewaan sepeda harian antar musim (spring, summer, fall, winter) selama periode tahun 2011-2012, yang dapat dijadikan dasar bagi tim operasional untuk menentukan jadwal pemeliharaan (maintenance) armada secara masif?
- **Pertanyaan 2:** Berapa besar selisih rata-rata penyewaan sepeda harian antara kondisi cuaca cerah dan cuaca buruk (hujan/salju) selama tahun 2011-2012, sebagai landasan bagi tim marketing untuk merancang strategi promo khusus di cuaca buruk?
- **Pertanyaan 3:** Berapa perbedaan rata-rata penyewaan sepeda harian antara hari kerja (workingday) dibandingkan dengan hari libur/akhir pekan selama periode tahun 2011-2012, agar tim marketing dapat merancang alokasi budget kampanye promo rekreasi yang tepat sasaran?

## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur tema visualisasi grafik agar lebih cantik
sns.set_theme(style='darkgrid')
```

## Data Wrangling

### Gathering Data

#### Load df_day

```python
# Memuat data penyewaan sepeda harian dari file CSV
df_day = pd.read_csv("day.csv")

# Menampilkan 5 baris pertama untuk memastikan data berhasil dimuat
df_day.head()
```

**Insight:** (Opsional)
- **Insight dari Gathering Data:**
- Telah berhasil memuat dataset penyewaan sepeda harian (`day.csv`) ke dalam bentuk DataFrame Pandas dengan nama variabel `df_day`.
- Berdasarkan tampilan awal, data berhasil terbaca dengan baik dan siap untuk dilanjutkan ke tahap pemeriksaan kelengkapan data (Assessing Data).

### Assessing Data

#### Identifying Data bike problem

```python
# 1. Memeriksa tipe data dan jumlah baris/kolom
print("Informasi Data:")
df_day.info()

print("\n" + "="*50 + "\n")

# 2. Memeriksa apakah ada data yang kosong (Missing Values)
print("Jumlah Missing Values:")
print(df_day.isna().sum())

print("\n" + "="*50 + "\n")

# 3. Memeriksa apakah ada data yang duplikat ganda
print("Jumlah Duplikasi Data: ", df_day.duplicated().sum())

print("\n" + "="*50 + "\n")

# 4. Memeriksa ringkasan statistik untuk melihat apakah ada angka yang aneh
print("Ringkasan Statistik:")
display(df_day.describe())
```

**Steps to Take:**
- **Steps to Take:**

* Mengubah tipe data pada kolom `dteday` dari yang awalnya `object` (string) menjadi tipe data `datetime` karena kolom tersebut berisi informasi tanggal.
* Mengubah (mapping) nilai angka pada kolom kategorikal seperti `season`, `weathersit`, dan `workingday` menjadi teks keterangan yang jelas agar nantinya mudah dibaca saat dibuatkan grafik.

**Insight:** (Opsional)
- **Insight: (Opsional)**

* Data terpantau sangat bersih dari segi kelengkapan. Tidak ditemukan adanya data yang kosong (*missing values*) maupun baris data yang duplikat pada dataset `day.csv`.
* Ditemukan ketidaksesuaian tipe data pada kolom `dteday` yang seharusnya merepresentasikan tanggal, namun masih terbaca sebagai teks biasa (`object`).
* Kolom yang merepresentasikan kategori (`season`, `weathersit`, `workingday`) masih menggunakan angka. Walaupun secara sistem tidak salah, hal ini akan membingungkan pembaca saat divisualisasikan nanti.

### Cleaning Data

#### Fixing Data bike problem

```python
# 1. Memperbaiki tipe data kolom dteday dari object menjadi datetime
df_day['dteday'] = pd.to_datetime(df_day['dteday'])

# 2. Mengubah angka pada kolom 'season' menjadi nama musim
season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
df_day['season'] = df_day['season'].map(season_mapping)

# 3. Mengubah angka pada kolom 'weathersit' menjadi keterangan cuaca
weather_mapping = {
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
}
df_day['weathersit'] = df_day['weathersit'].map(weather_mapping)

# 4. Mengubah angka pada kolom 'workingday' menjadi keterangan hari kerja/libur
workingday_mapping = {0: 'Holiday/Weekend', 1: 'Working Day'}
df_day['workingday'] = df_day['workingday'].map(workingday_mapping)

# Mengecek kembali informasi data untuk memastikan perubahan berhasil
df_day.info()
```

**Insight:** (Opsional)
**Insight dari Cleaning Data:**

* Telah dilakukan perbaikan tipe data pada kolom `dteday` menjadi `datetime` agar sistem mengenali kolom tersebut sebagai urutan waktu yang benar.
* Telah dilakukan proses *mapping* (pengubahan nilai) pada kolom `season`, `weathersit`, dan `workingday` dari format angka menjadi teks deskriptif. Hal ini bertujuan agar data jauh lebih mudah dipahami oleh pembaca awam saat dibuatkan grafik (*chart*) nantinya.
* Data *Bike Sharing* kini sudah sepenuhnya bersih, rapi, dan siap untuk diolah ke tahap Exploratory Data Analysis (EDA).

## Exploratory Data Analysis (EDA)

### Explore Data bike

```python
# Menampilkan statistik deskriptif
df_day.describe(include="all")

# 1. Eksplorasi Rata-rata Penyewaan Berdasarkan Musim (Menjawab Pertanyaan 1)
print("Rata-rata Penyewaan Sepeda Berdasarkan Musim (Season):")
display(df_day.groupby(by="season").agg({
    "cnt": ["mean", "max", "min"]
}).round())

print("\n" + "="*50 + "\n")

# 2. Eksplorasi Rata-rata Penyewaan Berdasarkan Cuaca (Menjawab Pertanyaan 2)
print("Rata-rata Penyewaan Sepeda Berdasarkan Kondisi Cuaca (Weathersit):")
display(df_day.groupby(by="weathersit").agg({
    "cnt": ["mean", "max", "min"]
}).round())

print("\n" + "="*50 + "\n")

# 3. Eksplorasi Rata-rata Penyewaan Berdasarkan Hari Kerja (Menjawab Pertanyaan 3)
print("Rata-rata Penyewaan Sepeda Berdasarkan Hari Kerja (Workingday):")
display(df_day.groupby(by="workingday").agg({
    "cnt": ["mean", "max", "min"]
}).round())
# 1. EDA Univariate: Melihat parameter statistik dari kolom numerik (jumlah penyewaan)
print("Distribusi data pada kolom jumlah penyewaan harian (cnt):")
display(df_day['cnt'].describe().round())

print("\n" + "="*50 + "\n")

# 2. EDA Univariate: Melihat distribusi data pada kolom kategorikal (jumlah hari tiap musim)
print("Jumlah hari pengamatan berdasarkan Musim:")
display(df_day['season'].value_counts())

# Melihat distribusi jumlah penyewaan
plt.figure(figsize=(10, 5))
sns.histplot(df_day['cnt'], kde=True)
plt.title("Distribusi Penyewaan Sepeda")
plt.show()

print("Explore Statistik Deskriptif:")
display(df_day.describe().T)
```

**Insight:** (Opsional)
**Insight dari Statistik Deskriptif (Univariate):**

* **Fluktuasi Penyewaan (`cnt`):** Rata-rata penyewaan sepeda harian berada di angka **4.504 unit**, rentang nilainya sangat ekstrem. Ada hari di mana penyewaan anjlok di **22 unit**, dan ada hari yang membludak hingga **8.714 unit**. Nilai standar deviasi yang tinggi (1.937) menandakan bahwa permintaan sangat tidak stabil dan bergantung pada faktor luar (seperti cuaca/musim).
* **Komposisi Pelanggan (`casual` vs `registered`):** Pelanggan yang sudah menjadi member (*registered*) adalah tulang punggung bisnis ini, dengan rata-rata **3.656 penyewaan/hari**. Angka ini jauh di atas pelanggan kasual yang hanya **848 penyewaan/hari**.
* **Kondisi Lingkungan:** Kelembapan (`hum`) harian rata-rata cukup tinggi, dan suhu (`temp`) memiliki variasi yang cukup besar. Faktor lingkungan ini sangat berpotensi menjadi dalang di balik naik-turunnya jumlah penyewaan.

## Visualization & Explanatory Analysis

### Pertanyaan 1: mengenai musim


```python
# Melakukan agregasi data untuk mencari rata-rata penyewaan per musim
season_rentals_df = df_day.groupby("season").cnt.mean().reset_index()

# Menampilkan tabel hasil hitungan untuk diperiksa
season_rentals_df
# Melakukan agregasi data untuk mencari rata-rata penyewaan per musim
season_rentals_df = df_day.groupby("season").cnt.mean().reset_index()

# Menampilkan tabel hasil hitungan untuk diperiksa
season_rentals_df
# Membuat visualisasi untuk Pertanyaan 1
plt.figure(figsize=(10, 6))

# Menentukan warna: Biru Tua untuk yang tertinggi, Biru Muda untuk lainnya
dark_blue = "#1565C0"
light_blue = "#90CAF9"
colors = [dark_blue if v == season_rentals_df.cnt.max() else light_blue for v in season_rentals_df.cnt]

ax = sns.barplot(
    x="season",
    y="cnt",
    data=season_rentals_df,
    palette=colors
)

# --- Penerapan Decluttering (Sesuai Materi) ---
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.set_ylabel(None)
ax.set_xlabel(None)
ax.tick_params(axis='y', left=False, labelleft=False)

# Menambahkan Data Labels di atas batang
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 10),
                textcoords='offset points',
                fontweight='bold', color='#333333')

plt.title("Rata-rata Penyewaan Sepeda Tertinggi Terjadi pada Musim Gugur (Fall)",
          loc="left", fontsize=14, fontweight='bold', pad=20)
plt.show()
```

**Insight:** (Opsional)
**Insight:**
**Insight:**
* Grafik batang di atas menunjukkan perbandingan rata-rata penyewaan sepeda harian untuk setiap musim. Dapat dilihat bahwa Musim Gugur (*Fall*) memiliki rata-rata penyewaan tertinggi dibandingkan musim lainnya.
* Sebaliknya, Musim Semi (*Spring*) tercatat sebagai musim dengan rata-rata penyewaan paling rendah selama periode 2011-2012.
* Terdapat perbedaan yang cukup signifikan antara jumlah penyewaan di Musim Semi dibandingkan dengan Musim Panas (*Summer*) dan Musim Gugur (*Fall*).

## Analisis Lanjutan (Opsional)

```python
# 1. Manual Clustering/Binning: Mengelompokkan hari berdasarkan performa penyewaan
# Kita bagi menjadi 3 kategori: Low, Medium, High
bins = [0, 2000, 5000, 9000]
labels = ['Low Rental Day', 'Medium Rental Day', 'High Rental Day']
df_day['rental_category'] = pd.cut(df_day['cnt'], bins=bins, labels=labels)

# Melihat distribusi jumlah hari pada tiap kategori
advanced_analysis = df_day.groupby('rental_category').agg({
    'cnt': ['count', 'mean', 'max']
}).round()

print("Distribusi Kategori Hari Penyewaan:")
display(advanced_analysis)

print("\n" + "="*50 + "\n")

# 2. Analisis Korelasi: Melihat hubungan suhu dan kelembapan terhadap jumlah penyewaan
correlation = df_day[['temp', 'atemp', 'hum', 'windspeed', 'cnt']].corr()
print("Korelasi Variabel Cuaca terhadap Jumlah Penyewaan (cnt):")
display(correlation['cnt'].sort_values(ascending=False))
```

**Insight:** (Opsional)
**Insight Analisis Lanjutan:**

* **Manual Clustering (Binning):** Dari hasil pengelompokan, mayoritas hari selama periode 2011-2012 masuk ke dalam kategori **Medium Rental Day** (2.000 - 5.000 penyewaan). Namun, jumlah hari yang masuk kategori **High Rental Day** (>5.000) juga cukup banyak. Ini menunjukkan bisnis memiliki potensi pasar yang besar dan sering mencapai titik puncak permintaan.
* **Analisis Korelasi:** Variabel **Suhu (temp/atemp)** memiliki korelasi positif yang kuat dengan jumlah penyewaan. Artinya, orang cenderung lebih suka menyewa sepeda saat cuaca terasa hangat. Sebaliknya, **Kelembapan (hum)** memiliki korelasi negatif, yang menandakan bahwa tingkat kelembapan yang terlalu tinggi (udara gerah/lengket) cenderung membuat orang malas bersepeda.
